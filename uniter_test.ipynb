```python
digit_index_mapping = {}
ordering = [12, -1, -1, 8, -1, 5, 9, 6, 1, -1, 4, 0, 3, 7, 2, 14, 15, -1, 10, 13, 11, -1]
for state in range(np.power(2,len(ordering))):
    converted_idx = uniter.convert_idx(ordering,state)
    print('state',state,'contributes to',converted_idx)
    if converted_idx in digit_index_mapping:
        digit_index_mapping[converted_idx].append(state)
    else:
        digit_index_mapping[converted_idx] = [state]
```
